```python
df_pos=pd.read_csv('긍정사전수정.csv')
df_neg=pd.read_csv('부정사전수정.csv')
df_pos
```

```python
pos_list=df_pos['index'].tolist()


neg_list=df_neg['index'].tolist()

print(pos_list)
```

```python
import re

pos_word=[]
for i in pos_list:
    word=re.sub('/[a-zA-Z]*','',i)
    pos_word.append(word)
pos_word

neg_word=[]
for i in neg_list:
    word=re.sub('/[a-zA-Z]*','',i)
    neg_word.append(word)
```

```python
pos_list=[]
for i in pos_word:
    word=i.split(';')
    for j in word:
        pos_list.append(j)
pos_list
```

```python
neg_list=[]
for i in neg_word:
    word=i.split(';')
    for j in word:
        neg_list.append(j)
neg_list
```

```python
score=[]
num=0
minutes_tone_list=[]
import re
for i in minutes_list:
    pos_st=0
    neg_st=0
    sentence_list=i.split('@@@')
    for sentence in sentence_list:
        numpos=0
        numneg=0
        for i in pos_list:
            if re.search(i,sentence):
                numpos+=1
        for j in neg_list:
            if re.search(j,sentence):
                numneg+=1        
        try:
            score_posneg=(numpos-numneg)/(numpos+numneg)
            score.append(score_posneg)
        except ZeroDivisionError:
            score_posneg=0
            score.append(score_posneg)
        if score_posneg>=0:
            pos_st+=1
        elif score_posneg<0:
            neg_st+=1
    print(pos_st,neg_st)
    minutes_tone=(pos_st-neg_st)/(pos_st+neg_st)
    minutes_tone_list.append(minutes_tone)
# # #     print(score)
#     for sc in score:
#         if sc>=0:
#             pos_st+=1
#         elif sc<0:
#             neg_st+=1
#     print(pos_st)
#     print(neg_st)
#     minutes_tone=(pos_st-neg_st)/(pos_st+neg_st)
#     print(minutes_tone)
#     minutes_tone_list.append(minutes_tone)
```
