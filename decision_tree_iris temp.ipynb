```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data[:,2:]
y = iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=2)
tree.fit(X,y)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y=iris.target

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=54)

from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=2,random_state=54)
tree.fit(X_train,y_train)

from sklearn.tree import export_graphviz
dot_data = export_graphviz(tree,out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True,filled=True,impurity=True)
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)

graph.write_png("iris.png")

from IPython.core.display import Image
Image("iris.png")
```
