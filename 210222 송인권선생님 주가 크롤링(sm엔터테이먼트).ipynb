```python
stocklists = []
for i in range(1,515):
    urls = 'https://finance.naver.com/item/sise_day.nhn?code=041510&page=' + str(i)
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36'}
    response = requests.get(urls, headers=headers)
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    tr = soup.select('table.type2 tr')
    for e in range(1,len(tr)-1):
        if tr[e].select('td')[0].text.strip():
            sdate = tr[e].select('td')[0].text.strip()
            sopen = tr[e].select('td')[1].text.strip()
            sclose = tr[e].select('td')[3].text.strip()
            shigh = tr[e].select('td')[4].text.strip()
            slow = tr[e].select('td')[5].text.strip()
            svol = tr[e].select('td')[6].text.strip()
            skey = tr[e].select('td')[0].text.replace(".","")
            stocklists.append([sdate,sopen,sclose,shigh,slow,svol,skey])
print(stocklists)
```

```python
#skey >> 원래 날짜가 유일한 값(key)인데 새로운 값을 추가할 가능성이 있어서 skey를 key(유일한 값, 인덱스로 보기 편하게 만든다.)로 만드려고 만듦
def data_process(d=[]):
    dbname = 'sm_ent.db'
    with sqlite3.connect(dbname) as conn: # db랑 파이썬이랑 연결시켜주기
        # conn = sqlite3.connect(dbname) >> 원래 쓰는 방식
        cur = conn.cursor() #데이터 베이스에 커서 가져다 놓기(손을 대려고 하는 거에요)
        if d: #d안에 무엇이 들어있는지 확인하기(있으면 통과, 아니면 else)
            sql = 'select skey from sm_ent_주가 where skey={}'.format(d[6]) #d[6] = skey
            # spl >> skey 값을 sm_ent_주가에서 찾아라(기존에 값이 있는지 확인(유일한 값인지))
            check_key = cur.execute(sql).fetchall()
            # sql에 모든 skey 값 찾기 실행
            if check_key:
                print("중복있음",d[6]) #있으면 '중복있음' 프린트
            else : 
                sql = "insert into sm_ent_주가 (sdate,sopen,sclose,shigh,slow,svol,skey) values(?,?,?,?,?,?,?)"
                cur.execute(sql,d) #없으면 새로운 값 데이터베이스에 추가
                print('입력완료')
        else : #select = 조회
            sql = 'select * from sm_ent_주가' # 이전에 만든 데이터베이스를 보여주고 끝내라
            result = cur.execute(sql).fetchall()
            print(result)
```

```python
for s in stocklists:
    data_process(s)
```
