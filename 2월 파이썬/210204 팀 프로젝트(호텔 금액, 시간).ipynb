```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
X_train = pd.read_csv('X_train.csv', encoding='cp949')
y_train=pd.read_csv('y_train.csv', encoding='cp949')
train = pd.merge(X_train, y_train, on='custid', how='left')
amount = train.groupby('part').sum()
# amount = amount.transpose()
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount["."] = amount[""]+amount[""]
# amount = amount.drop(["","","","",
#                              "","","","",
#                              "","","","",
#                              "",""],axis=1)
# amount = amount.transpose()
amount = amount.reset_index()
###  , 
fig = plt.figure(figsize=(15,15))
plt.rc('font', family='Malgun Gothic', size=20)
labels=amount['part']
colors = sns.color_palette('hls',len(labels))
patches, texts, autotexts = plt.pie(amount['amount'],
                                    labels=amount['part'],
                                    startangle=90, #
                                    counterclock=False, # 
                                    autopct='%.2f%%', #pi  
                                    colors = colors,
                                    pctdistance=0.7, #pi   
                                    labeldistance=1.05 #lable pi 
#                                     rotatelabels=True
                                   )
### label 
for t in texts:
    t.set_color("black")
    t.set_fontsize(20)
### label  
for t in autotexts:
    t.set_color("white")
    t.set_fontsize(20)
###   
# centre_circle = plt.Circle((0,0),0.3,color='white')
# plt.gca().add_artist(centre_circle)
title = 'part  '
plt.title(title, size=40)
plt.show()
```
